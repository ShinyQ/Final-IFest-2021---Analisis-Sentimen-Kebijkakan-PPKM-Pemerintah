```python
import pandas as pd
from sklearn.utils import shuffle

def balancing_df(path=None):
  if path == None:
    raise "No data"

  df = pd.read_csv(path)
  minimum = [len(df[df["label"] == 2]), len(df[df["label"] == 1]), len(df[df["label"] == 0])]
  minimum = min(minimum)
  df_0 = df[df["label"] == 0].sample(minimum)
  df_1 = df[df["label"] == 1].sample(minimum)
  df_2 = df[df["label"] == 2].sample(minimum)

  df_gabung = pd.concat([df_0, df_1, df_2])
  df_gabung = shuffle(df_gabung)
  df_gabung = df_gabung.reset_index(drop=True)

  return df_gabung
  
```
